```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sn
sn.set() # overrides the matplotlib style of plotting graph
```

```python
df = pd.read_csv("1.01.+Simple+linear+regression.csv")
df.describe()
```

```python
y = df['GPA']
x1 = df['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize = 40)
plt.ylabel('GPA', fontsize = 40)
plt.show()
```

```python
# y = bo + b1x
# before prediction we need to find the bo and b1 variables
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.2750
fig = plt.plot(x1, yhat, lw = 4, c='red', label = 'linear regression')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1
fig = plt.plot(x1, yhat, lw = 4, c = 'green', label = 'linear regression')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.xlim(0)
plt.ylim(0)
plt.show()
```

```python
plt.scatter(x1,y)
yhat = 0*x1 + 0.275
fig = plt.plot(x1,yhat, lw=4, c='red', label ='regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```
