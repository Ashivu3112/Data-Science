```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import statsmodels.api as sm
sn.set()
```

```python
df = pd.read_csv('real_estate_price_size_year.csv')
df.describe()
```

```python
df
```

```python
y = df['price']
x1 = df[['size', 'year']]
x1
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
df2 = pd.read_csv('real_estate_price_size_year_view.csv')
df2
```

```python
df2.describe(include = 'all')
```

```python
df3 = df2.copy()
df3['view'] = df3['view'].map({'Sea view': 1, 'No sea view': 0})
df3.head()
```

## Create the regression

```python
y = df3['price']
x1 = df3[['size','year','view']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
